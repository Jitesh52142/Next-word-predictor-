```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
## we go sentense by sentenses :
```

```python
for sentence in data.split('\n'):
    print(sentence)
```

```python
# we here use tokinizer to goinig throw word bu word 
```

```python
input_sequences = []

for sentence in data.split('\n'):
    tokenized_sample=(tokenizer.texts_to_sequences([sentence])[0])

    for i in range(1,len(tokenized_sample)):
        input_sequences.append(tokenized_sample[:i+1])
```

```python
input_sequences
```

```python
## we use zero padding to equalize the list shape 
```

```python
max_len =max([len(x) for x in input_sequences])# len of max string
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
y
```

```python
y
```

```python
## it is a multi class classification task 
```

```python
# then we do one how encoding 
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y,num_classes=221)#always geeting one no higher on num_class 
```

```python
y.shape
```

```python
y
```

```python
## now we create a artitacture 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()

model.add(Embedding(input_dim=221,output_dim=100,input_length=27))
model.add(LSTM(150))
model.add(Dense(221,activation='softmax'))

```

```python
model.build(input_shape= (None ,27))
```

```python
model.compile(loss='categorical_crossentropy',optimizer ='adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
## now we train or model 
```

```python
model.fit(x,y,epochs = 100,validation_data=(x,y))
```
